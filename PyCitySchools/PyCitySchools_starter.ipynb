## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Total number of schools
Total_schools = len(school_data_complete["school_name"].unique())

#Total number of students
Total_students = len(school_data_complete["Student ID"].value_counts())

#Total budget of all schools
Budget_unique = school_data_complete["budget"].unique()
Total_budget = Budget_unique.sum()

#Average scores in maths and reading
Average_maths = school_data_complete["maths_score"].mean()
Average_reading = school_data_complete["reading_score"].mean()

#Percentage of scores
Maths_pass = school_data_complete[school_data_complete.iloc[:,6] >= 50]
Total_maths_pass = len(Maths_pass["Student ID"].value_counts())
Maths_pass_percentage = Total_maths_pass/Total_students

Reading_pass = school_data_complete[school_data_complete.iloc[:,5] >= 50]
Total_reading_pass = len(Reading_pass["Student ID"].value_counts())
Reading_pass_percentage = Total_reading_pass/Total_students

#Overall pass percentage
Overall_pass = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
Total_overall_pass = len(Overall_pass["Student ID"].value_counts())
Overall_pass_percentage = Total_overall_pass/Total_students

Area_summary = pd.DataFrame({"Total Schools": [Total_schools],
    "Total Students": Total_students,
    "Total Budget": Total_budget,
    "Average Maths Score": Average_maths,
    "Average Reading Score": Average_reading,
    "% Passing Maths": Maths_pass_percentage,
    "% Passing Reading": Reading_pass_percentage,
    "% Overall Passing": Overall_pass_percentage})

#Use Map to format
Area_summary["Total Budget"] = Area_summary["Total Budget"].map("${:,.2f}".format)
Area_summary["Total Students"] = Area_summary["Total Students"].map("{:,}".format)
Area_summary["% Passing Maths"] = (Area_summary["% Passing Maths"]*100).map("{:.6f}%".format)
Area_summary["% Passing Reading"] = (Area_summary["% Passing Reading"]*100).map("{:.6f}%".format)
Area_summary["% Overall Passing"] = (Area_summary["% Overall Passing"]*100).map("{:.6f}%".format)
Area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Grouping by school

School_name_group = school_data_complete.groupby(['school_name'])

#School types - Govt/Independant
School_type = school_data.set_index(["school_name"])["type"]
#Total students in a school
School_total_students = School_name_group["Student ID"].count()
#Per student budget in each school
School_total_budget = School_name_group['budget'].first()
Per_student_budget = School_total_budget/School_total_students

#Average scores in each school
SchAvg_math_score = School_name_group["maths_score"].mean()
SchAvg_reading_score = School_name_group["reading_score"].mean()

#% passing scores in maths, reading and overall
Group_passing_math = school_data_complete[school_data_complete["maths_score"]>=50].groupby(["school_name"])["Student ID"].count()
Percent_passing_math = (Group_passing_math/School_total_students)*100
Group_passing_reading = school_data_complete[school_data_complete["reading_score"]>=50].groupby(["school_name"])["Student ID"].count()
Percent_passing_reading = (Group_passing_reading/School_total_students)*100
Overall_passing = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('school_name')['Student ID'].count()
Percent_passing_overall = (Overall_passing/School_total_students)*100

# Create a DataFrame
Schools_summary = pd.DataFrame({
    "School Type": School_type,
    "Total Students": School_total_students,
    "Total School Budget": School_total_budget,
    "Per Student Budget": Per_student_budget,
    "Average Maths Score": SchAvg_math_score,
    "Average Reading Score": SchAvg_reading_score,
    "% Passing Maths": Percent_passing_math,
    "% Passing Reading": Percent_passing_reading,
    "% Overall Passing": Percent_passing_overall
})
Schools_summary

# Create a copy of school summary data frame before formatting to be able to use the numeric data on original data frame later, else we get error
Copy_school_summary = Schools_summary.copy()

#Use Map to format
Copy_school_summary["Total School Budget"] = Schools_summary["Total School Budget"].map("${:,.2f}".format)
Copy_school_summary["Per Student Budget"] = Schools_summary["Per Student Budget"].map("${:,.2f}".format)

Copy_school_summary


```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins
Spending_bins = [0, 584.99, 614.99, 644.99, 675]
Group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Create a new data frame by locating the desired columns
Scores_spending = Schools_summary.loc[:,['Average Maths Score',
                                  'Average Reading Score','% Passing Maths',
                                  '% Passing Reading','% Overall Passing']]
# Add a new columns named Spending Ranges (Per Student) and binning based off budget per student
Scores_spending['Spending Ranges (Per Student)']= pd.cut(Schools_summary["Per Student Budget"],Spending_bins,labels=Group_names)

# Create a group based off of the bins
Scores_spending = Scores_spending.groupby('Spending Ranges (Per Student)').mean()
Scores_spending
```
